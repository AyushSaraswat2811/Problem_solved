```python
# Import necessary libraries
import pandas as pd


data_file_path = "Leads.csv"  
df = pd.read_csv(data_file_path)

# Display the first 5 rows
print("ðŸ”¹ First 5 rows of the dataset:")
display(df.head())

# Check the structure of the dataset
print("\nðŸ”¹ Dataset Information:")
df.info()

# Check missing values in each column
print("\nðŸ”¹ Missing Values in Each Column:")
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
display(missing_data[missing_data['Missing Values'] > 0])

# Check unique values in categorical columns
print("\nðŸ”¹ Unique Values in Categorical Columns:")
for col in df.select_dtypes(include=['object']).columns:
    print(f"{col}: {df[col].nunique()} unique values")

# Check the target variable distribution
print("\nðŸ”¹ Target Variable (Converted) Distribution:")
display(df['Converted'].value_counts(normalize=True) * 100)  # Percentage of Converted (1) and Not Converted (0)
```
